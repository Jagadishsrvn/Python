### Enumerate Function
    >> The enumerate() function adds a counter to an iterable (like a list, tuple, or string) and returns an enumerate object containing pairs of (index, value). It simplifies looping over iterables while tracking both the index and the element.

```python
string = 'python'

for char in enumerate(string):
    print(char)
```

```python
string = 'python'

for index, value in enumerate(string):
     print(f"Character at index {index}: {value}")
```

```python
text=input("Enter a text:")
vowel="aeiouAEIOU"
for index,char in enumerate(text):
    if char in vowel:
        print(f"Vowel '{char}' at index '{index}'")
```

# f-string formatting
    >> F-strings (formatted string literals) are a way to embed variables, expressions, and function calls directly inside a string using { }.

```python
name = "John"  
age = 25  
print(f"Hello, {name}! You are {age} years old.")
```

```python
a = 5  
b = 3  
print(f"{a} + {b} = {a + b}") 
```

# dot formatting
    >>the format() method is a powerful and flexible way to format strings. It allows you to insert variables or values into placeholders defined by curly braces {}

```python
name = "John"
age = 30
message = "My name is {} and I am {} years old.".format(name, age)
print(message)
print("My name is {} and I am {} years old".format(name, age))
```

### String functions/methods


## 1. capitalize()
    >>> It will capitalize only first char of string, other char will be in small case first cchar (zero index char string).
### Syntax:
    variable_name.capitalize()

```python
# 1.
text = "hello world"
result = text.capitalize()
print(result)
```

```python
# 2.
text = "python programming"
new_text = text.capitalize()
print(text)
print(new_text)
```

```python
# 3.
print("Enter the name:")
for i in range(5):
    name = input(f"Enter name {i+1}: ")
    print(name.capitalize())
```

## 2. lower()
    >> It return all string in lower case
### Syntax: 
    variable_name.lower() 

```python
# 1.
text = "HELLO WORLD"
result = text.lower()
print(result)
```

```python
# 2.
str1 = "Python"
str2 = "PYTHON"

if str1.lower() == str2.lower():
    print("The strings are equal (case-insensitive).")
else:
    print("The strings are not equal.")
```

## 2. upper()
    >> It return all string in upper case
### Syntax: 
    variable_name.upper() 

```python
# 1.
text = "python"
new_text = text.upper()
print(text)       
print(new_text)
```

```python
# 2.
str1 = "Python"
str2 = "PYTHON"

if str1.upper() == str2.upper():
    print("The strings are equal (case-insensitive).")
else:
    print("The strings are not equal.")
```

## 4. title()
     >> It will convert string into title case capitilize first char of every word
### Syntax: 
    variable_name.title()

```python
# 1.
text = "hello world"
result = text.title()
print(result)
```

```python
# 2.
text = "pYtHoN pRoGrAmMiNg"
result = text.title()
print(result)
```

```python
# 3.
str1 = "Python"
str2 = "PYTHON"

if str1.title() == str2.title():
    print("The strings are equal (case-insensitive).")
else:
    print("The strings are not equal.")
```

## 5. lstrip()
    >> It will remove leading white spaces & returns left trim version string
### Syntax:
    variable.lstrip()

```python
# 1.
text = "   Hello, Python!"
result = text.lstrip()
print(result)
```

```python
# 2. 
s1 = '           python'
print(s1)
print(len(s1))
print(s1.lstrip())
```

```python
# 2.
text = "###python###"
result = text.lstrip("#")
print(len(text))
print(result)
```

## 6. rstrip()
    >>> It will remove trailing white spaces & returns right trim version string
### Syntax:
    string_name.rstrip()

```python
# 1.
text = "   Hello, Python!                   "
result = text.rstrip()
print(result)
```

```python
# 2. 
s3 = 'python           '
print(s3)
print(len(s3))
print(s3.rstrip())
print(len(s3.rstrip()))
```

```python
s1 = '        python        '
print(s1.lstrip().rstrip().upper())
```

```python
s1 = '        python        '
print(s1.lstrip().rstrip().title())
```
