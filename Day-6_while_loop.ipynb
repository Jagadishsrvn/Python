## While loop: 
    >> A while loop repeats a block of code as long as a condition is True. It’s used when you don’t know the number of iterations in advance
#### Syntax
    while condition:  
        # Code to repeat 
    #Used for unknown iterations (e.g., user input).

```python
#1.Print a number from 1 tp 10 using while loop

count=1
while count<=10:
    print(count)
    count=count+1 #Increment to avoid infinite loop
print("Done counting!")
```

```python
#2.Write a program to print first 10 natural numbers in reveerse order

count=10
while count >0:
    print(count)
    count=count-1 #Decrement to avoid infinite loop
print("Completed")
```

```python
#4.Write a program to find the factorial of a number
num=int(input("Enter a number:"))
fac=1
while(num>0):
    fac=fac*num
    print(fac)
    num-=1
print("Factorial is :",fac)
    
```

```python
#5.Write a program to reverse string
text = input("Enter a string")

# Initialize variables
reversed_text = ""
text_len = len(text) - 1

# While loop to reverse the string
while text_len >= 0:
    reversed_text += text[text_len]
    text_len -= 1
    # Output the reversed string
print("Reversed string:", reversed_text)
```

```python
#7.Write a program to check if a number is a palindrome using a while loop.
num = int(input("Enter a number: "))
original_num = num
reversed_num = 0
while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num = num // 10
if original_num == reversed_num:
    print("It's a palindrome!")
else:
    print("It's not a palindrome.")
```

```python
#8. Use a while loop to find the greatest common divisor (GCD) of two numbers.
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
while b != 0:
    temp = b
    b = a % b
    a = temp
print("The GCD is:", a)
```
