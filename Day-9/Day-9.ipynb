## 8. replace()
    >> It will return a string where a substing is replaced with that value
### Syntax:
    string.replace(old_substring, new_substring, occurances)

```python
text = "Ram and Raju went to the market, while coming back to the home Raju brought a pen"
result = text.replace("Raju", "Rajesh")
print(result)
```

```python
text = "Ram and Raju went to the market, while coming back to the home Raju brought a pen"
result = text.replace("Raju", "Rajesh",1)
print(result)
```

### 9. count()
    >> It will return the number of times a specified substring occurs in string
### Syntax:
    string.count('substring',[start_index], [end_index])

```python
text = "Python is easy. Python is powerful."
count_python = text.count("Python")
print(f"'Python' appears {count_python} times.")
```

```python
# Count occurrences of a word
text = "Python is easy. Python is powerful."
count_python = text.count("Python",0,10)

print(f"'Python' appears {count_python} times.")
```

### 23. zfill()
    >> The zfill() method pads the beginning of a string with leading zeros to reach a specified width. If the original string's length is already greater than or equal to the width, it returns the original string.
#### Syntax:
    string.zfill(width)

```python
t="567"
print(len(t))
print(t.zfill(10))
```

```python
text1 = "123"
print(text1.zfill(5))
```

```python
text2 = "hello"
print(text2.zfill(3)) #length is already greater
```

### 25. center()
    The center() method returns a centered string of a specified width. The original string is centered, and the remaining space is filled with a specified fill character (default is a space).
#### Syntax:
        string.center(width, fillchar)

```python
text = "Python"
print(text.center(10))       
print(text.center(10, "*"))    
print(text.center(15, "-"))
```
