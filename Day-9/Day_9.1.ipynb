```python
#1. Write a program that takes a string as input and prints its lowercase, uppercase, and title-cased versions.
text = input("Enter a string: ")
lowercase_text = text.lower()
uppercase_text = text.upper()
titlecase_text = text.title()
print(f"Lowercase: {lowercase_text}")
print(f"Uppercase: {uppercase_text}")
print(f"Title Case: {titlecase_text}")
```

```python
# 2. Write a program that takes a string with leading and trailing spaces and prints the string with those spaces removed.
text_with_spaces = "   Hello, World!   "
stripped_text = text_with_spaces.strip()
print(f"Original: '{text_with_spaces}'")
print(f"Stripped: '{stripped_text}'")
```

```python
# 3. Write a program that takes a sentence and replaces all occurrences of a specific word (you can choose the word) with another word.
sentence = "The cat sat on the mat."
old_word = "cat"
new_word = "dog"
new_sentence = sentence.replace(old_word, new_word)
print(f"Original sentence: {sentence}")
print(f"New sentence: {new_sentence}")
```

```python
# 4. Write a program that takes a string and a character as input and counts how many times that character appears in the string (case-insensitive).
text = input("Enter a string: ")
char_to_count = input("Enter a character to count: ").lower()
count = 0
for char in text.lower():
    if char == char_to_count:
        count += 1
print(f"The character '{char_to_count}' appears {count} times.")
```

```python
# 5. Write a program that takes a string and a substring as input and prints the index of the first occurrence of the substring. If not found, print -1.
text = input("Enter a string: ")
substring_to_find = input("Enter a substring to find: ")
index = text.find(substring_to_find)
print(f"The substring '{substring_to_find}' is found at index: {index}")
```
