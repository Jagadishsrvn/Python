### list functions/methods

#### 1. append
    >> Adds an item to the end of the list.
####   Syntax:
    list.append(item)

```python
fruits = ["apple", "banana"] 
print(fruits)
fruits.append("cherry")  
print(fruits)
```

```python
var1 = [3,4,5]
var2 = [10,8,9] # this will be considered as single item
y =var1.append(var2)
print(var1)
print(len(var1))
```

#### 2. extend()
    >>> Adds all elements of an iterable (e.g., list, tuple) to the end.
#### Syntax:
    list.extend(iterable)

```python
list1 = [1, 2, 3]
list2 = [4, 5]
list1.extend(list2)
print(list1)
```

#### 3. insert()
    >> When we want to add item at particulat index
#### Syntax:
    list.insert(index,item)

```python
numbers = [10, 20, 40]  # 
numbers.insert(2, 30)  # Insert 30 at index 2  
print(numbers)
```

```python
fruits = ['apple', 'banana', 'cherry']
fruits.insert(1, 'orange')
print(fruits)
```

#### 4. remove()
    >> Removes the first occurrence of the item.
#### Syntax:
    list.remove(item)

```python
data = [5, 2, 3, 2]  
data.remove(2)  
print(data)
```

```python
var2= [3, 4, 5, 'data',10,10.5]
var2.remove('data')
var2
```

```python
#remove item "Math" from the list
sub_names = ['Phy', 'Che', 'Math', 'Eng', 'Bio']
val = 'Math'
if val in sub_names:
    sub_names.remove(val)
    print("item Deleted")
else:
    print("item not found")
```

#### 5. pop()
    >> Removes and returns the item at index (default: last item).

```python
fruits = ["apple", "banana", "cherry"]  
popped = fruits.pop(1)# Remove "banana"
print(popped) 
poppp=fruits.pop(0) # remove apple
print(poppp)
```

```python
ls=[1,2,4,5,67,786]
ls.pop(1)
ls
ls.pop() #default last item is removed
ls
```

#### 6. clear()
    >> Removes all items from the list.

```python
items = [1, 2, 3]  
items.clear()  
print(items)
```

#### 7. sort()
    >> Sorts the list in-place (ascending by default).
#### Syntax:
    list.sort() #for ascending order
    list.sort(reverse=True) #for descending order

```python
# Sort in ascending order
numbers = [3, 1, 4, 2]  
numbers.sort()  
print(numbers)  
```

```python
# Sort in descending order  
numbers.sort(reverse=True)  
print(numbers)
```

```python
# 7. Create a list of 5 names. Print each name in uppercase using a for loop.
str_list = []
n = int(input("How many strings do you want to add? "))
for i in range(n):
    word = input(f"Enter string {i+1}: ")
    str_list.append(word.upper())
print("Final list:", str_list)
```

```python
# 10 Write a program to filter out odd and even numbers from the list, create two list containing odd and even numbers.
ls=[12,4,23,56,67,87,87,98]
odd_num=[]
even_num=[]
for i in ls:
    if i % 2==0:
        even_num.append(i)
    else:
        odd_num.append(i)
print(f"odd numbers : {odd_num}")
print(f"even numbers : {even_num}")
```
