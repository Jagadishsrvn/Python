### String

    >> Used to represent text data (e.g., names, messages, file paths).
    >> String can contains digit(0-9), space, special character(!@#$%^&*) inside the quotes                                     
    >> Immutable: Once created, strings cannot be modified (but new strings can be created).                                     
    >> Strings are used to represent textual data

### Indexing in String
    >> It is used to access only one character/particular characters from string.
    >> Positive index will always start from zero (0).Negative indexing is also allowed (starts from -1 for the last character).
    >> Use [] to print particular string item.
                    P  Y  T  H  O  N
                    0  1  2  3  4  5  -> Positive indexing
                   -6 -5 -4 -3 -2 -1  -> Negative indexing

```python
# Positive index
word="PYTHON"
print(word[0])
print(word[1])
print(word[2])
print(word[3])
print(word[4])
print(word[5])
```

```python
# Negative index
word="PYTHON"
print(word[-6])
print(word[-5])
print(word[-4])
print(word[-3])
print(word[-2])
print(word[-1])
```

```python
# String reverse order using positive indexing
word="PYTHON"
print(word[5])
print(word[4])
print(word[3])
print(word[2])
print(word[1])
print(word[0])
```

```python
# String reverse order using negative indexing
word="PYTHON"
print(word[-1])
print(word[-2])
print(word[-3])
print(word[-4])
print(word[-5])
print(word[-6])
```

### Slicing
    >> Slicing allows you to extract a substring (portion) of a string by specifying a range of indices.
    >> Default start index is 0.
    >> Default end index is len(string)
    Syntax:
        string[start_index(include) : end_index[excluded] : [step_size]]

```python
# Slicing using positive indexing
string='python'
print(string[0:6])
print(string[0:5])
print(string[0:4])
print(string[0:3])
print(string[0:2])
print(string[0:1])
```

```python
# Slicing using negative indexing
string='python'
print(string[-1:-7:-1])
print(string[-1:-6:-1])
print(string[-1:-5:-1])
print(string[-1:-4:-1])
print(string[-1:-3:-1])
print(string[-1:-2:-1])
```

```python
string='python'
print(string[0:7:1])
# Printing alternative letters using slicing
print(string[0:7:2])
```

```python
slicing="Slicing allows you to extract a substring (portion) of a string by specifying a range of indices."
# Finding the lenght of string
lenght=len(slicing)
print(lenght)
# Printing alternative letters using slicing
print(slicing[0:98:2])
```

```python
# Assignment Operators: Used to assign values to variables. -> =,+=,-=,*=,/=
# 1.= (Assign):
x = 5
y=10
print(x,y)
# ----------------------------------------------------------------------------------------------------------------------------------------------------
# 2. += (Add AND assign): x += 3 is equivalent to x = x + 3
x+=3
print(x)
# -----------------------------------------------------------------------------------------------------------------------------------------------------
# 3. -= (Subtract AND assign): x -= 3 is equivalent to x = x - 3
x-=5
print(x)
# -----------------------------------------------------------------------------------------------------------------------------------------------------
# 4. *= (Multiply AND assign): x *= 3 is equivalent to x = x * 3
x*=10
print(x)
# -----------------------------------------------------------------------------------------------------------------------------------------------------
# 5. /= (Divide AND assign): x /= 3 is equivalent to x = x / 3
x/=6
print(x)
```

```python
#Comparison (Relational) Operators: Used to compare two values and return a True or False (Boolean) result.
# 1. == (Equal to): x == y
a=5
b=5
print("Output :",a == b)
c=10
d=25
print("Output :",c==d)
```

```python
# 2. != (Not equal to): x != y
a=10
b=25
print("Output :",a!=b)
c=5
d=5
print("Output :",c != d)
```

```python
# 3. > (Greater than): x > y
a = 15
b = 10
print("Output :",a > b)
```

```python
# 5. >= (Greater than or equal to): x >= y

a = 10
b = 10
print("Output :",a >= b)
c = 20
d = 10
print("Output :",c >= d)
c = 2
d = 10
print("Output :",c >= d)
```

```python
# Logical Operators: Used to combine conditional statements.
# 1. and operator : Returns True if both statements are true.
a = 10
b = 20

# using and operator
if a > 5 and b > 15:
    print("Both conditions are True")
else:
    print("At least one condition is False")

```

```python
# 2. or operator : Returns True if at least one of the statements is true
x = 5
y = 15

# using or operator
if x > 10 or y > 10:
    print("At least one condition is True")  # 
else:
    print("Both conditions are False")
```

```python
# 3. not: Reverses the result, returns False if the result is true.
age = 18
if not age <18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```
