```python
# Write a program that takes three sides of a triangle as input and determines the type of triangle:
# If all three sides are equal, print "Equilateral Triangle".
# If two sides are equal:
# If the third side is greater than the sum of the other two, print "Not a Triangle".
# Otherwise, print "Isosceles Triangle".
# If no sides are equal:
# If the sum of any two sides is less than or equal to the third side, print "Not a Triangle".
# Otherwise, print "Scalene Triangle".
#-------------Solution------------------------
s1=int(input("Enter A side cm value:"))
s2=int(input("Enter B side cm value:"))
s3=int(input("Enter C side cm value:"))
if s1==s2==s3:
    print("Equilateral Triangle")
else:
    if s1==s2 or s2==s3 or s3==s1:
        if s1==s2:
            if s3>s1+s2:
                print("Not a Triangle")
            else:
                print("Isosceles Triangle")
        elif s2==s3:
            if s1>s3+s2:
                print("Not a Triangle")
            else:
                print("Isosceles Triangle")
        else:
            if s2>s1+s3:
                print("Not a Triangle")
            else:
                print("Isosceles Triangle")
    else:
        if(s1+s2<=s3) or (s2+s3<=s1) or (s1+s3<=s2):
            print("Not a Triangle")
        else:
            print("Scalene Triangle")
```

```python
# Write a program to calculate the final price of a product based on the following conditions:

# If the price is greater than ₹5000:
# If the customer is a member, apply a 20% discount.
# Otherwise, apply a 10% discount.
# If the price is between ₹1000 and ₹5000:
# If the customer is a member, apply a 15% discount.
# Otherwise, apply a 5% discount.
# If the price is below ₹1000, no discount is applied.
# -----------------Solution--------------------------
price=int(input("Enter the amount:"))
print("Total bill amount is :",price)
member=input("Do you have membership here ? \'Yes' or \'No'")
if price>5000:
    if member.title() == "Yes":
        print("Apply 20% discount")
        print("Discount amount is:",20/100*price)
        print("Final amount is:",price-(20/100*price))
    else:
        print("Apply 10% discount")
        print("Discount amount is:",10/100*price)
        print("Final amount is:",price-(10/100*price))
elif price>=1000 and price<=5000:
    if member.title() == "Yes":
        print("Apply 15% discount")
        print("Discount amount is:",15/100*price)
        print("Final amount is:",price-(15/100*price))
    else:
        print("Apply 5% discount")
        print("Final amount is:",price-(5/100*price))

else:
    print("No discount is applies")
```
