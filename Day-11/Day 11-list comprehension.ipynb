### 1. Generate squares of numbers from 0 to 9 using list comprehension

```python
square_list=[i**2 for i in range(0,10)]
print(square_list)
```

### 2.Create a list containing cubes from 11 to 20

```python
# Without List Comprehension (Traditional Loop)
cube_list=[]
for i in range(11,21):
    cube_list.append(i**3)
print(cube_list)
```

```python
# Using list comprehension 
cube_list1=[i**3 for i in range(11,21)]
print(cube_list1)
```

### 3. Create new list using existing list such that new list have all characters in upper case

```python
# Without List Comprehension (Traditional Loop)
fruit = ['Kiwi', 'banana', 'orange', 'apple', 'dragonfruit', 'cherry']
new_fruit_list=[]
for i in fruit:
    new_fruit_list.append(i.upper())
print(new_fruit_list)
```

```python
# Using list comprehension
new_fruit_list=[i.upper() for i in fruit]
print(new_fruit_list)
```

### Categorize Numbers as "Even" or "Odd"

```python
# Without List Comprehension (Traditional Loop)
numbers = [31, 42, 63, 84, 65]
labels=[]
for i in numbers:
    if i%2==0:
        labels.append("even".title())
    else:
        labels.append("odd".title())
print(labels)
```

```python
# Using list comprehension
numbers = [31, 42, 63, 84, 65]
labels=["even".title() if i % 2==0 else "odd".title() for i in numbers]
print(labels)
```

### Filtering positive numbers from a list

```python
# without using list comprehenstion:
numbers1 = [-2, -1, 0, 1, 2, 3]
pos_number=[]
neg_number=[]
for i in numbers1:
    if i > 0:
        pos_number.append(i)
    else:
        neg_number.append(i)
print(pos_number)
print(neg_number)
```

```python
# Using list comprehension
numbers1 = [-2, -1, 0, 1, 2, 3]
num1=[i for i in numbers1 if i >0]
print(num1)
```
